# **IPL Data analysis**

This dataset consist about 8 years IPL matches records and 18 columns associated with each of them.
> The 9 columns are as follows:-


> id : Match number

> season : The year in which the match happened

> city : The city where the match happened

> date : Match time

> team1 : Home team

> team2 : Away team

> tosswinner : Which team wins the toss

> tossdicision : Which decided by winning the toss

> result : Normal or Not normal

> dl_applied : Duckworth-Lewis-Stern applied or not

> winner : The team that won

> win_by_runs : The team that bowled won

> win_by_wickets : The team that batted won

> player_of_the_match : Who is the player of the match

> venue : Stadium name

> umpire1 : Field umpire

> umpire2 : Field umpire

> umpire3 : Third umpire name

**Extracting the records where a team won batting first**

```python
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
# Looking at head
batting_first.head()
```

```python
# Making a Histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of runs")
plt.xlabel("Runs")
plt.show()
```

**Finding out the number of wins each team after batting first**

```python
batting_first['winner'].value_counts()
```

Chennai Super Kings have won by batting first the most times

**Top 3 successfull batting first team**

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["yellow","blue","purple"])
plt.title("Top 3 successfull batting first team")
plt.show()
```

Top 3 successfull batting first teams are : 


> 1) Chennai Super Kings 


> 2) Mumbai Indians 


> 3) Kolkata Knight Riders







```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

Chennai Super Kings have the highest batting first win percentage among all the teams.
